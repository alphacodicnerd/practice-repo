```python
import numpy as np
import pandas as pd
from numpy.random import randn
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
```

```python
address = 'train.csv'

df = pd.read_csv(address, index_col='Order Date', parse_dates=True, encoding='utf8')
df.head()
```

## Starting Seaborn - Creating stats data graphics

```python
import seaborn as sb
sb.set_style('whitegrid')
```

```python
address = "/Users/tanyasrivastava/Documents/Python For Data Science Practice notebooks/practice-repo/mtcars.csv"
cars = pd.read_csv(address)

cars.columns = ['cars_name', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.head()
cars.index = cars.cars_name

mpg = cars['mpg']
mpg.plot(kind='hist')
```

```python
plt.hist(mpg) #diffrent method
plt.plot()
```

```python
sb.distplot(mpg)
```

### Scatterplots

```python
cars.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150)
```

```python
#With seaborn

sb.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

```python
# scatterplot matrix
cars_subset = cars[['mpg','disp','hp','wt']]
sb.pairplot(cars_subset)
plt.show()
```

### Building Boxplot

```python
cars.boxplot(column='mpg',by='am')
cars.boxplot(column='wt',by='am')
```

```python
sb.boxplot(x='am',y='mpg',data=cars, palette='hls')
```
